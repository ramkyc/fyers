```python
df['tick_time'] = pd.to_datetime(df['tick_time'])
#df['tick_time'] = df['tick_time'] + pd.to_timedelta('5h30min')
#print(df['tick_time'].dtype)
print(df['tick_time'])
```

```python

# Filter for the specific symbol and time range
df = df[(df['trading_symbol'] == 'NIFTY24O2425000PE') &
        (df['tick_time'] >= '2024-10-18 03:45:00+00:00') &
        (df['tick_time'] <= '2024-10-18 10:00:00+00:00')]

# Convert UTC time to IST
df['tick_time'] = df['tick_time'] + pd.to_timedelta('5h30min')

# Group by 15-minute intervals and calculate the open price
df['15min_interval'] = df['tick_time'].dt.floor('15min')
df['15min_open'] = df.groupby('15min_interval')['u_open'].transform('first')

# Create a new column to indicate if the tick price crossed above the 15-minute open
df['crossed_above'] = (df['u_ltp'] > df['15min_open']) & (df['u_ltp'].shift(1) <= df['15min_open'])

# Count the number of times the price crossed above for each 15-minute interval
crossovers = df.groupby('15min_interval')['crossed_above'].sum()

print(crossovers)
```

```python
# Filter for the specific symbol and time range
df_filtered = df[(df['trading_symbol'] == 'NIFTY24O2425200CE') &
                (df['tick_time'] >= '2023-12-31 09:30:00') &
                (df['tick_time'] <= '2023-12-31 15:30:00')]

# Check if there's any data for the filtered criteria
if len(df_filtered) > 0:
  # Proceed with calculations if data exists
  df_filtered['tick_time'] = pd.to_datetime(df_filtered['tick_time'])
  df_filtered['tick_time_ist'] = df_filtered['tick_time'] + pd.to_timedelta('5h30min')
  df_filtered['15min_interval'] = df_filtered['tick_time_ist'].dt.floor('15T')
  df_filtered['15min_open'] = df_filtered.groupby('15min_interval')['u_open'].transform('first')
  df_filtered['crossed_above'] = (df_filtered['u_ltp'] > df_filtered['15min_open']) & (df_filtered['u_ltp'].shift(1) <= df_filtered['15min_open'])
  crossovers = df_filtered.groupby('15min_interval')['crossed_above'].sum()
  print(crossovers)
else:
  print("No data found for the specified criteria")
```
