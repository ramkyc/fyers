```python
# Calculating high-low, open-low and open-high ranges for the candles.
df_17OCTOptions.loc[:, 'hlr'] = df_17OCTOptions['high'] - df_17OCTOptions['low']
df_17OCTOptions.loc[:, 'olr'] = df_17OCTOptions['open'] - df_17OCTOptions['low']
df_17OCTOptions.loc[:, 'ohr'] = df_17OCTOptions['high'] - df_17OCTOptions['open']
```

```python
# Calculating 5 ema for the calculated candle ranges above.
df_17OCTOptions.loc[:, 'hlr_ema_5'] = df_17OCTOptions.loc[:, 'hlr'].ewm(span=5, adjust=False, min_periods=5).mean()
df_17OCTOptions.loc[:, 'olr_ema_5'] = df_17OCTOptions.loc[:, 'olr'].ewm(span=5, adjust=False, min_periods=5).mean()
df_17OCTOptions.loc[:, 'ohr_ema_5'] = df_17OCTOptions.loc[:, 'ohr'].ewm(span=5, adjust=False, min_periods=5).mean()
```

```python
# For the current row open, SL is previous row olr_ema
df_17OCTOptions.loc[:, 'sl'] = df_17OCTOptions['open'] - df_17OCTOptions.loc[:, 'olr_ema_5'].shift(1)

# Target is: current row open + 1.5* previous row olr_ema
df_17OCTOptions.loc[:, 'target'] = df_17OCTOptions.loc[:, 'open'] + (df_17OCTOptions.loc[:, 'olr_ema_5'].shift(1))*1.2
#df_17OCTOptions['target'] = df_17OCTOptions['target'].round(2)

Tgtcount = (df_17OCTOptions['high'] >= df_17OCTOptions['target']).sum()
print("Number of times target achieved in the same candle:", Tgtcount)

SLcount = (df_17OCTOptions['low'] <= df_17OCTOptions['sl']).sum()
print("Number of times SL hit in the same candle:", SLcount)

# How to count if the target or sl is hit in the subsequent candles?
def count_target_achievements(df_17OCTOptions):
    """Counts the instances where the 'target' value is achieved in subsequent rows.

    Args:
        df (pd.DataFrame): The DataFrame containing the data.

    Returns:
        int: The count of target achievements.
    """

    target_achieved = False
    count = 0

    for index, row in df_17OCTOptions.iterrows():
        if row['high'] >= row['target'] and not target_achieved:
            target_achieved = True
            count += 1
        elif row['high'] < row['target']:
            target_achieved = False

    return count

# Example usage
count = count_target_achievements(df_17OCTOptions)
print("Number of target achievements in sebsequent candles:", count)
```
